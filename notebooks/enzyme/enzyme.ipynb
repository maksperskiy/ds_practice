### Chars

```python
chars = []
```

```python
df["protein_sequence"].apply(set).apply(
    lambda x: [chars.append(el) for el in x if el not in chars]
);
```

```python
chars = sorted(chars)
```

```python
print(chars)
```

### Encoding

```python
encoding = pd.get_dummies(chars)
```

```python
encoding
```

```python
chars_df = pd.DataFrame(df["protein_sequence"].apply(list).tolist())
```

```python
df.shape
```

```python
chars_df.head()
```

```python
chars_df.shape
```

```python
encoding["nan"] = 0
```

```python
encoding = encoding.apply(lambda x: x.apply(bool), axis=1)
```

```python
def encode_char(char):
    if not char:
        char = "nan"
    return encoding[char].tolist()


def encode_row(row):
    return pd.Series(row.apply(encode_char).tolist())
```

```python
multiindex = np.hstack(
    np.array(
        [
            np.linspace(
                0, chars_df.shape[1], len(chars) * chars_df.shape[1], dtype=int, endpoint=False
            ).reshape(-1, 1),
            np.array(chars * chars_df.shape[1]).reshape(-1, 1),
        ]
    )
)
```

```python
multiindex.shape
```

```python
multiindex
```

```python
%%time
encoded_df = chars_df.apply(encode_row, axis=1)
```

```python
%%time
encoded_df = pd.concat(
    (
        [
            pd.DataFrame(encoded_df.pop(col).tolist()).add_prefix(f"number_{col}_")
            for col in encoded_df.columns
        ]
    ),
    axis=1,
)
```

```python
%%time
encoded_df = pd.DataFrame(encoded_df.values, columns=multiindex)
```

```python
%%time
encoded_df.columns = pd.MultiIndex.from_tuples(
    encoded_df.columns, names=["Number", "Char"]
)
```

```python
encoded_df
```

```python
encoded_df.to_pickle("data/enzyme/encoded_df.pkl")
```
